```python
import pandas as pd
import psycopg2
from sqlalchemy import create_engine
```

```python
# Ganti nilai berikut sesuai dengan informasi koneksi PostgreSQL Anda
db_user = 'postgres'
db_password = '7288'
db_host = 'localhost'
db_port = '5432'
db_name = 'SimpleDataEngineering'
```

```python
# Path ke file CSV
csv_path = 'ElectronicsProductsPricingData.csv'

# Membaca file CSV menjadi DataFrame menggunakan pandas
df = pd.read_csv(csv_path)

df 
```

```python
# Membuat koneksi ke PostgreSQL menggunakan psycopg2
conn = psycopg2.connect(
    user=db_user,
    password=db_password,
    host=db_host,
    port=db_port,
    database=db_name
)
```

```python
# Membuat kursor untuk eksekusi perintah SQL
cursor = conn.cursor()

# Menyiapkan perintah SQL untuk membuat tabel (jika belum ada)
create_table_query = """
CREATE TABLE IF NOT EXISTS electronics_product (
    id VARCHAR,
    prices_amountMax FLOAT,
    prices_amountMin FLOAT,
    prices_availability VARCHAR,
    prices_condition VARCHAR,
    prices_currency VARCHAR,
    prices_dateSeen VARCHAR,
    prices_isSale BOOLEAN,
    prices_merchant VARCHAR,
    prices_shipping VARCHAR,
    prices_sourceURLs VARCHAR,
    asins VARCHAR,
    brand VARCHAR,
    categories VARCHAR,
    dateAdded VARCHAR,
    dateUpdated VARCHAR,
    ean VARCHAR,
    imageURLs VARCHAR,
    keys VARCHAR,
    manufacturer VARCHAR,
    manufacturerNumber VARCHAR,
    name VARCHAR,
    primaryCategories VARCHAR,
    sourceURLs VARCHAR,
    upc VARCHAR,
    weight VARCHAR
);
"""

cursor.execute(create_table_query)
conn.commit()

```

```python
# Menggunakan SQLAlchemy untuk mengupload DataFrame ke PostgreSQL
engine = create_engine(f'postgresql://{db_user}:{db_password}@{db_host}:{db_port}/{db_name}')
df.to_sql('electronics_product', engine, if_exists='replace', index=False)

# Menutup koneksi
cursor.close()
conn.close()

print("Data berhasil diupload ke PostgreSQL.")
```

```python
# melihat nilai nul dari masing-masing kolom
# Menggunakan loop untuk melihat persentase nilai null pada masing-masing kolom
for column in df.columns:
    null_percentage = df[column].isnull().mean() * 100
    print(f"Persentase nilai null pada kolom {column}: {null_percentage:.2f}%")
```

```python
# melihat unique value pada kolom kategori

# Menggunakan loop untuk melihat unique value pada masing-masing kolom
for column in df.columns:
    if df[column].dtype == 'object':
        unique_values = df[column].unique()
        print(f"Unique values for column {column}:")
        print(unique_values)
        print('\n')
```

```python
# Mengambil kolom-kolom numerik
numeric_columns = df.select_dtypes(include=['float64', 'int64'])

# Menampilkan rentang nilai pada masing-masing kolom
for column in numeric_columns.columns:
    min_value = numeric_columns[column].min()
    max_value = numeric_columns[column].max()
    print(f"Rentang nilai pada kolom {column}: {min_value} - {max_value}")
```
